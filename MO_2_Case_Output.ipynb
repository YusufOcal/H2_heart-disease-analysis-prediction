### **Veri Seti Sütunları ve Açıklamaları**

1. **`id`**:
   - Hastaya ait benzersiz kimlik numarası.
   - Her bir satırı temsil eden bir tanımlayıcıdır.

2. **`age`**:
   - Hastanın yaşı (yıl olarak).

3. **`sex`**:
   - Hastanın cinsiyeti.
     - `1`: Erkek
     - `0`: Kadın

4. **`dataset`**:
   - Veri seti kaynağını veya grubunu temsil eder.
   - Bu sütun veri setinin birleştirildiği farklı alt grupları ifade eder.

5. **`cp`** (Chest Pain Type - Göğüs Ağrısı Tipi):
   - Hastanın yaşadığı göğüs ağrısı türünü ifade eder.
     - `1`: Tipik angina (kalp kaynaklı göğüs ağrısı)
     - `2`: Atipik angina (kalp dışı nedenlerle göğüs ağrısı)
     - `3`: Non-anginal pain (kalp dışı göğüs ağrısı)
     - `4`: Asymptomatic (semptomsuz)

6. **`trestbps`** (Resting Blood Pressure - Dinlenme Kan Basıncı):
   - Hastanın istirahat halindeki kan basıncı (mmHg cinsinden).

7. **`chol`** (Serum Cholesterol - Serum Kolesterol Seviyesi):
   - Hastanın serum kolesterol seviyesi (mg/dL).

8. **`fbs`** (Fasting Blood Sugar - Açlık Kan Şekeri):
   - Hastanın açlık kan şekeri seviyesi.
     - `1`: Eğer açlık kan şekeri > 120 mg/dL ise
     - `0`: Diğer durumlar

9. **`restecg`** (Resting Electrocardiographic Results - Dinlenme Elektrokardiyografi Sonuçları):
   - EKG sonuçlarının kategorik olarak değerlendirilmesi.
     - `0`: Normal
     - `1`: ST-T dalga anormallikleri (T dalgası inversiyonu veya ST yükselmesi/depresyonu)
     - `2`: Sol ventrikül hipertrofisi belirtileri

10. **`thalch`** (Maximum Heart Rate Achieved - Maksimum Kalp Atış Hızı):
    - Hastanın egzersiz sırasında ulaştığı maksimum kalp atış hızı.

11. **`exang`** (Exercise Induced Angina - Egzersiz Sonrası Angina):
    - Egzersiz sonrası göğüs ağrısı olup olmadığını belirtir.
      - `1`: Evet
      - `0`: Hayır

12. **`oldpeak`**:
    - ST depresyonu değeri (dinlenme ile egzersiz sonrası arasındaki fark).
    - Kalp hastalığı riskini değerlendirmek için kullanılan bir ölçüm.

13. **`slope`** (Slope of the Peak Exercise ST Segment - Egzersiz ST Segmentinin Eğimi):
    - ST segmentinin egzersiz sonrası eğimini ifade eder.
      - `1`: Yukarı eğimli
      - `2`: Düz
      - `3`: Aşağı eğimli

14. **`ca`** (Number of Major Vessels Colored by Fluoroscopy - Floroskopi ile Görüntülenen Büyük Damar Sayısı):
    - Kalp damarlarının daralma derecesine bağlı olarak floroskopi ile renklendirilen damar sayısı (0–3 arası değerler alır).

15. **`thal`** (Thalassemia - Talasemi Durumu):
    - Hastanın talasemi durumunu belirtir.
      - `3`: Normal
      - `6`: Sabit defekt
      - `7`: Tersine döndürülebilir defekt

16. **`num`** (Diagnosis of Heart Disease - Kalp Hastalığı Teşhisi):
    - Kalp hastalığının varlığını gösteren hedef değişken.
      - `0`: Kalp hastalığı yok
      - `1, 2, 3, 4`: Kalp hastalığının ciddiyet derecesini gösterir (1 en az, 4 en ciddi).

````python
def sayisal_sutunlari_doldur(df):
    """
    Sayısal sütunları seçip eksik değerleri (NaN) sütunun ortalamasıyla doldurur.

    Args:
    df (pd.DataFrame): Veri çerçevesi

    Returns:
    pd.DataFrame: Eksik değerleri doldurulmuş veri çerçevesi
    """
    # Sayısal sütunları seç
    sayisal_sutunlar = df.select_dtypes(include=['int64','float64' ]).columns

    # Eksik değerleri doldur
    for sutun in sayisal_sutunlar:
        if df[sutun].isnull().sum() > 0:  # Eksik (NaN) değer kontrolü
            df[sutun].fillna(df[sutun].mean(), inplace=True)

    return df
````

````python
def kategorik_sutunlari_doldur(df):
    """
    Kategorik sütunları seçip eksik değerleri (NaN) mod (en sık görülen değer) ile doldurur.

    Args:
    df (pd.DataFrame): Veri çerçevesi

    Returns:
    pd.DataFrame: Eksik değerleri doldurulmuş veri çerçevesi
    """
    # Kategorik sütunları seç
    kategorik_sutunlar = df.select_dtypes(include=['object', 'category']).columns

    # Eksik değerleri doldur
    for sutun in kategorik_sutunlar:
        if df[sutun].isnull().sum() > 0:  # Eksik (NaN) değer kontrolü
            df[sutun].fillna(df[sutun].mode()[0], inplace=True)

    return df
````

#### 3. Aykırı Değer Analizi
   - **Box Plot ile Görselleştirme**: Her sayısal sütun için box plot çizerek aykırı değerleri görselleştirin.
   - **Aykırı Değerlerin Tespiti**: IQR yöntemiyle aykırı değerleri belirleyin. Q1 (1. çeyrek) ve Q3 (3. çeyrek) değerlerine göre alt ve üst sınırları hesaplayarak, bu sınırların dışında kalan verileri tespit edin.
   - **Aykırı Değerlerin İşlenmesi**: Aykırı değerleri alt ve üst sınırlarla değiştirerek baskılayın veya bu veri noktalarını veri setinden çıkarın.

````python
def detect_outliers(dataframe, column, q1=0.05, q3=0.95): #q1 ve q3 değerleri çok uçtan aldık.
    # İlk ve üçüncü çeyrek değerlerini hesaplama
    Q1 = dataframe[column].quantile(q1)
    Q3 = dataframe[column].quantile(q3)
    IQR = Q3 - Q1
    # Alt ve üst sınırları belirleme
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    # Aykırı değerleri filtreleme
    outliers = dataframe[(dataframe[column] < lower_limit) | (dataframe[column] > upper_limit)]
    return outliers, lower_limit, upper_limit
````

 Bu fonksiyon, bir veri çerçevesindeki bir sütunda **aykırı değerleri** belirlemek için **IQR (Interquartile Range)** yöntemini kullanır. Fonksiyonun detaylı açıklaması aşağıda verilmiştir:

---

### **detect_outliers Fonksiyonun Amacı**
Fonksiyon, bir veri çerçevesindeki belirli bir sütunda **aykırı değerleri** tespit etmek için **çeyrek değerler (quartiles)** ve **IQR** yöntemini kullanır. Kullanıcı, sütunun hangi yüzdelerle çeyrek değerlerin hesaplanacağını belirleyebilir.

---

### **Parametreler**
1. **`dataframe`**:
   - Aykırı değerlerin tespit edilmek istendiği veri çerçevesi.

2. **`column`**:
   - Aykırı değerlerin inceleneceği sütunun adı.

3. **`q1` ve `q3`**:
   - **q1**: İlk çeyrek değerin hesaplanacağı yüzde (varsayılan: 0.05 yani %5).
   - **q3**: Üçüncü çeyrek değerin hesaplanacağı yüzde (varsayılan: 0.95 yani %95).
   - Bu parametrelerle kullanıcı, aykırı değer tespitinde kullanılan sınırları daraltabilir ya da genişletebilir.

---

### **Fonksiyonun Çalışma Mantığı**
1. **Çeyrek Değerlerin Hesaplanması**:
   - `Q1`: Verilen sütundaki `q1` yüzdelik değeri (örn. %5) bulunur.
   - `Q3`: Verilen sütundaki `q3` yüzdelik değeri (örn. %95) bulunur.

2. **IQR Hesaplama**:
   - **IQR** (Interquartile Range): Çeyrekler arası fark hesaplanır.
     \[
     \text{IQR} = Q3 - Q1
     \]

3. **Alt ve Üst Sınırların Belirlenmesi**:
   - Aykırı değerlerin alt ve üst sınırlarını belirlemek için şu formüller kullanılır:
     \[
     \text{Alt Sınır} = Q1 - 1.5 \times \text{IQR}
     \]
     \[
     \text{Üst Sınır} = Q3 + 1.5 \times \text{IQR}
     \]

4. **Aykırı Değerlerin Filtrelenmesi**:
   - Sütundaki değerler `lower_limit`'ten küçük veya `upper_limit`'ten büyükse aykırı değer olarak değerlendirilir.
   - Aykırı değerler şu şekilde filtrelenir:
     ```python
     outliers = dataframe[(dataframe[column] < lower_limit) | (dataframe[column] > upper_limit)]
     ```

---

### **Fonksiyonun Dönüş Değerleri**
Fonksiyon, aşağıdaki üç değeri döndürür:
1. **`outliers`**:
   - Tespit edilen aykırı değerlerin olduğu satırları içeren bir DataFrame.

2. **`lower_limit`**:
   - Alt sınır (bir değerin bundan küçük olması aykırı olduğunu gösterir).

3. **`upper_limit`**:
   - Üst sınır (bir değerin bundan büyük olması aykırı olduğunu gösterir).

---

### **Fonksiyonun Esneklikleri**

- **Genel Kullanım**:

Bu şekilde, bir sütunun aykırı değerlerini esnek bir şekilde tespit edebilirsiniz.

````python
# Sayısal sütunlar için aykırı değerlerin adetini bulalım:

for col in df.select_dtypes(include=['number']).columns:
    outliers = detect_outliers(df, col)
    print(f"{col} sütununda {len(outliers[0])} aykırı değer bulundu.")
````

````python
# bir dataframede aykırı değer olup olmadığını kontrol edelim:

def has_outliers(dataframe, column, q1=0.05, q3=0.95):

    _, lower_limit, upper_limit = detect_outliers(dataframe, column, q1, q3)
    return dataframe[(dataframe[column] < lower_limit) | (dataframe[column] > upper_limit)].shape[0] > 0
````

Bu fonksiyonun amacı, bir sütunda **aykırı değerler olup olmadığını** kontrol etmektir. Bunun için **`detect_outliers`** fonksiyonunu kullanır ve sınırları kontrol ederek **boolean bir sonuç (True/False)** döndürür.

---

### **has_outliers Fonksiyonun Çalışma Mantığı**
```python
def has_outliers(dataframe, column, q1=0.05, q3=0.95):
    _, lower_limit, upper_limit = detect_outliers(dataframe, column, q1, q3)
    return dataframe[(dataframe[column] < lower_limit) | (dataframe[column] > upper_limit)].shape[0] > 0
```

#### **Parametreler**
1. **`dataframe`**:
   - Kontrol edilecek veri çerçevesi.
   
2. **`column`**:
   - Aykırı değerlerin olup olmadığını kontrol etmek istediğiniz sütunun adı.
   
3. **`q1` ve `q3`**:
   - Çeyrek değerleri belirleyen yüzdelikler (varsayılan: %5 ve %95).

---

#### **Fonksiyonun Çalışma Adımları**
1. **`detect_outliers` Fonksiyonunu Çağırır**:
   - `detect_outliers` fonksiyonu ile çeyrek değerler (Q1 ve Q3), **alt sınır** ve **üst sınır** hesaplanır.
   - İlgili sütun için bu sınırların dışında kalan değerler tespit edilir.

   ```python
   _, lower_limit, upper_limit = detect_outliers(dataframe, column, q1, q3)
   ```
   - **`lower_limit`**: Alt sınır.
   - **`upper_limit`**: Üst sınır.

2. **Aykırı Değer Kontrolü**:
   - Aykırı değer olup olmadığını kontrol etmek için şu filtreleme yapılır:
     ```python
     dataframe[(dataframe[column] < lower_limit) | (dataframe[column] > upper_limit)]
     ```
     - `dataframe[column] < lower_limit`: Alt sınırın altındaki değerler.
     - `dataframe[column] > upper_limit`: Üst sınırın üstündeki değerler.
     - `|`: Bu iki durumu "veya" mantığıyla birleştirir.

3. **Sonucu Döndürme**:
   - Aykırı değerlerin toplam sayısını bulmak için:
     ```python
     dataframe[(dataframe[column] < lower_limit) | (dataframe[column] > upper_limit)].shape[0]
     ```
     - `shape[0]`: Bu filtreye uyan satırların (aykırı değerlerin) sayısını verir.

   - Eğer aykırı değerlerin sayısı sıfırdan büyükse, fonksiyon **`True`** döndürür; aksi halde **`False`** döner:
     ```python
     return dataframe[(dataframe[column] < lower_limit) | (dataframe[column] > upper_limit)].shape[0] > 0
     ```

---

### **Fonksiyonun Döndürdüğü Değer**
- **`True`**:
  - Eğer ilgili sütunda **en az bir aykırı değer** varsa.

- **`False`**:
  - Eğer ilgili sütunda **hiç aykırı değer yoksa**.

````python
# aykırı değerleri, alt sınır veya üst sınır ile baskılamak istersek:

def replace_outliers_with_limits(dataframe, column, lower_quantile=0.05, upper_quantile=0.95):

    _, lower_limit, upper_limit = detect_outliers(dataframe, column, lower_quantile, upper_quantile)
    dataframe[column] = dataframe[column].apply(lambda x: lower_limit if x < lower_limit else (upper_limit if x > upper_limit else x))
    print(f"{column} sütunundaki aykırı değerler {lower_limit} ve {upper_limit} sınırlarına göre değiştirilmiştir.")
````

---

### **replace_outliers_with_limits Fonksiyonun Amacı**
Bir veri çerçevesindeki bir sütundaki aykırı değerleri tamamen kaldırmak yerine, onları sınır değerleriyle (alt veya üst) değiştirerek veri kaybını önler. Bu yöntem, özellikle makine öğrenmesi modellerine daha düzgün ve kararlı veriler sağlamak için kullanılır.

---

### **Parametreler**
1. **`dataframe`**:
   - İşlem yapılacak veri çerçevesi.

2. **`column`**:
   - Aykırı değerlerin tespit edileceği ve değiştirileceği sütun.

3. **`lower_quantile` ve `upper_quantile`**:
   - Aykırı değerlerin hesaplanmasında kullanılan çeyrek değer yüzdelikleri.
   - Varsayılan olarak `0.05` (%5) ve `0.95` (%95) alınır.

---

### **Fonksiyonun Çalışma Mantığı**
1. **`detect_outliers` Fonksiyonunu Çağırır**:
   - `detect_outliers` fonksiyonu, sütundaki alt ve üst sınırları belirler:
     ```python
     _, lower_limit, upper_limit = detect_outliers(dataframe, column, lower_quantile, upper_quantile)
     ```
   - **`lower_limit`**: Sütundaki değerlerin alt sınırı.
   - **`upper_limit`**: Sütundaki değerlerin üst sınırı.

2. **Aykırı Değerleri Sınırlarla Değiştirme**:
   - Sütundaki her bir değeri kontrol eder ve aykırı olanları uygun sınır değerle değiştirir:
     ```python
     dataframe[column] = dataframe[column].apply(
         lambda x: lower_limit if x < lower_limit else (upper_limit if x > upper_limit else x)
     )
     ```
   - **Lambda Fonksiyonu**:
     - Eğer bir değer `lower_limit`'ten küçükse, bu değer **`lower_limit`** ile değiştirilir.
     - Eğer bir değer `upper_limit`'ten büyükse, bu değer **`upper_limit`** ile değiştirilir.
     - Diğer tüm değerler olduğu gibi kalır.

3. **Bilgilendirme**:
   - İşlemin sonucunda, hangi sütunun hangi sınır değerlerine göre değiştirildiği belirtilir:
     ```python
     print(f"{column} sütunundaki aykırı değerler {lower_limit} ve {upper_limit} sınırlarına göre değiştirilmiştir.")
     ```

---

### **Fonksiyonun Çıktısı**
Fonksiyon, sütundaki aykırı değerleri değiştirir ve veri çerçevesini günceller. İşlem sonunda bir açıklama mesajı yazdırır.

---

### **Örnek Kullanım**
```python
import pandas as pd

# Örnek bir DataFrame
data = {'value': [10, 12, 15, 18, 30, 50, 100, 150]}
df = pd.DataFrame(data)

# Aykırı değerleri sınırlarla değiştir
replace_outliers_with_limits(df, 'value', lower_quantile=0.05, upper_quantile=0.95)

# Sonuç
print(df)
```

#### **Örnek Veri Öncesi**
```plaintext

```

#### **İşlem**
1. **Alt ve Üst Sınırların Hesaplanması**:
   - `lower_limit`: 8.1
   - `upper_limit`: 44.7

2. **Değerlerin Güncellenmesi**:
   - `100` ve `150` değerleri **`upper_limit`** olan `44.7` ile değiştirilir.

#### **Sonuç Veri**
```plaintext

```

- **Makine öğrenmesi** modellerinde, aykırı değerlerin aşırı etkisini azaltmak için.
- Veri kaybını önlemek için aykırı değerleri tamamen kaldırmak yerine sınır değerlerle değiştirme.
- Düzensiz dağılımları daha düzgün hale getirmek için.
