```python
# FONKSİYON TANIMLANSAYDI:

def check_missing_data_and_dtypes(df):

    """
    Verilen DataFrame'deki her bir sütunun eksik veri sayısını ve veri tiplerini döndürür.

    Bu fonksiyon, her sütun için eksik değerlerin sayısını ve veri tiplerini hesaplar ve
    bu bilgileri birleştirerek döndüren bir DataFrame oluşturur.

    Parametreler:
    df (pandas.DataFrame): İncelenecek olan pandas DataFrame.

    Returns:
    pandas.DataFrame: Her bir sütunun eksik veri sayısı ve veri tipini içeren bir DataFrame.

    Örnek:
    result = check_missing_data_and_dtypes(df)
    """

    missing_data = df.isnull().sum()
    data_types = df.dtypes

    result = pd.concat([missing_data, data_types], axis=1) # birleştirme
    result.columns = ['Eksik Veri Sayısı', 'Veri Tipi'] # sütun isimlerini belirtme

    return result # çıktı

check_missing_data_and_dtypes(df) #fonksiyonu çağırma işlemi
```

```python
  # FONKSİYON OLARAK YAZILSAYDI:

def fill_missing_values(df):
    """
    Verilen DataFrame'deki eksik (NaN) değerleri doldurur.

    Bu fonksiyon, her sütunun veri tipine göre eksik değerleri uygun şekilde doldurur:
    - Sayısal sütunlar (float64, int64): Eksik değerler sütunun ortalaması ile doldurulur.
    - Kategorik sütunlar (object): Eksik değerler sütunun en sık görülen değeri (mod) ile doldurulur.

    Parametreler:
    df (pandas.DataFrame): Eksik verilerle dolu olan ve eksik verileri doldurulacak DataFrame.

    Returns:
    pandas.DataFrame: Eksik verileri doldurulmuş olan DataFrame.

    Örnek:
    df = fill_missing_values(df)
    """

    for sutun in df.columns:
        if df[sutun].dtype in ('float64', 'int64'):
            df[sutun] = df[sutun].fillna(df[sutun].mean())
            #df[sutun].fillna(df[sutun].mean(), inplace=True)
        if df[sutun].dtype == 'object': # 'category', 'boolean' gibi veriler de eklenebilirdi.
            df[sutun] = df[sutun].fillna(df[sutun].mode()[0])
            #df[sutun].fillna(df[sutun].mode()[0], inplace=True)
    return df
```

# Aykırı Verilerin Tespiti ve Giderilmesi

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sayısal sütunları seçmek
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
```

```python

plt.figure(figsize=(16, 6))  # Grafik boyutunu ayarlama
sns.boxplot(data=df[numeric_columns]) # Grafiği çizdiren komut

plt.title('Boxplot: Sayısal Sütunlar ve Aykırı Değerler', fontsize=14) # Başlık
plt.xticks(rotation=90)  # Sütun isimlerini dikey hale getirme
plt.show()
```

```python

# Aykırı değerleri tespit etmek için IQR analizi
for column in numeric_columns:

    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Aykırı değerlerin alt ve üst sınırları
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Aykırı değerleri tespit et
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    # Aykırı değerlerin sayısını yazdır
    print(f"{column} - Aykırı Değer Sayısı: {outliers.shape[0]}")
```

```python
  # FONKSİYONLU HALİ

def detect_outliers(dataframe, column, q1=0.25, q3=0.75): #q1 ve q3 değerleri çok uçtan aldık.

    # İlk ve üçüncü çeyrek değerlerini hesaplama
    Q1 = dataframe[column].quantile(q1)
    Q3 = dataframe[column].quantile(q3)
    IQR = Q3 - Q1

    # Alt ve üst sınırları belirleme
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    # Aykırı değerleri filtreleme
    outliers = dataframe[(dataframe[column] < lower_limit) | (dataframe[column] > upper_limit)]
    return outliers, lower_limit, upper_limit
```

```python
def replace_outliers(dataframe, column, q1=0.25, q3=0.75):
    """
    Belirtilen sütundaki aykırı değerleri tespit eder ve bunları alt ve üst limitlere göre düzeltir.

    Aykırı değerler, IQR (Interquartile Range) ile belirlenir. Eğer sütunda aykırı değer varsa,
    bu değerler alt ve üst çeyreklerin limitlerine çekilir.

    Parametreler:
    dataframe (pandas.DataFrame): İşlem yapılacak pandas DataFrame.
    column (str): Aykırı değerlerin tespit edileceği sütun adı.
    q1 (float, optional): İlk çeyrek değeri. Varsayılan 0.25.
    q3 (float, optional): Üçüncü çeyrek değeri. Varsayılan 0.75.

    Dönen Değer:
    pandas.DataFrame: Aykırı değerleri düzeltilmiş DataFrame.
    """

    # Aykırı değerleri tespit et
    outliers, lower_limit, upper_limit = detect_outliers(dataframe, column, q1, q3)

    if not outliers.empty:
        # Aykırı değerleri alt ve üst limitlere çekme
        dataframe[column] = dataframe[column].apply(
            lambda x: lower_limit if x < lower_limit else (upper_limit if x > upper_limit else x)
        )

    return dataframe
```

```python
# Sayısal sütunlar için aykırı değerleri düzeltme işlemi
for sutun in numeric_columns:
    # Aykırı değerleri düzelt
    df = replace_outliers(df, sutun)
```

```python
import seaborn as sns

# Her sayısal sütun için box plot çizimi
plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(len(numeric_columns), 1, i)
    sns.boxplot(data=df, x=col)
    plt.title(f"{col} - Box Plot")
    plt.xlabel(col)
plt.tight_layout()
plt.show()
```
